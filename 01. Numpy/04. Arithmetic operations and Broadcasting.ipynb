# **What is Numpy?**
**NumPy** là thư viện cơ bản cho tính toán khoa học bằng Python. Nó là một thư viện Python cung cấp một đối tượng mảng đa chiều, các đối tượng dẫn xuất khác nhau (chẳng hạn như các mảng và ma trận có mặt nạ) và một loạt các quy trình cho các hoạt động nhanh trên mảng, bao gồm toán học, logic, thao tác hình dạng, sắp xếp, lựa chọn, I / O , các phép biến đổi Fourier rời rạc, đại số tuyến tính cơ bản, các phép toán thống kê cơ bản, mô phỏng ngẫu nhiên và hơn thế nữa.

# **Arithmetic operations**
Tất cả các toán tử số học thông thường (`+`, `-`, `*`, `/`, `//`, `**`, etc.) đều có thể sử dụng được với **ndarrays**

Các phép toán dưới đây vận hành trên mảng cùng hình dạng, mỗi phần tử mảng này sẽ thực hiện phép toán với phần tử tương ứng mảng kia

```python
a = np.array([14, 23, 32, 41])
b = np.array([5, 4, 3, 2])
```

```python
print("a + b =", a + b)
```

```python
print("a - b =", a - b)
```

```python
print("a * b =", a * b)
```

```python
print("a / b =", a / b)
```

```python
print("a // b =", a // b)
```

```python
print("a % b =", a % b)
```

```python
print("a ** b =", a ** b)
```

Các mảng phải có hình dạng giống nhau. Nếu không, **NumPy** sẽ áp dụng các quy tắc **Broadcasting**

# **Broadcasting**
Tổng quát, khi các mảng không cùng hình dạng, **Numpy** sẽ áp dụng quy tắc **Broadcasting**

### **First rule**
Nếu các mảng không có cùng **rank**, thì 1 sẽ được thêm vào trước các mảng có **rank** nhỏ hơn cho đến khi **rank** của chúng khớp với nhau.

```python
arr = np.arange(10)
arr
```

```python
arr = arr.reshape(1, 1, 10)
arr
```

Thử cộng mảng 1D với mảng **arr** (3D)

```python
arr + np.arange(0, 100, 10)
```

### **Second rule**

```python
arr2 = np.arange(9).reshape(3, 3)
arr2
```

```python
arr2 + [[100], [200], [300]]
```

Áp dụng **First rule** và **Second rule**

```python
arr2 + [1000, 2000, 3000]
```

```python
arr2 + 500
```

### **Upcasting**
Khi cố gắng kết hợp các mảng với các kiểu **dtypes** khác nhau, **NumPy** sẽ **upcast** một kiểu có khả năng xử lý tất cả các giá trị có thể có

```python
k1 = np.arange(-3, 8, dtype=np.int8)
k1
```

```python
k2 = k1 + np.arange(0, 11, dtype=np.uint8)
k2
```

```python
k3 = k1 + 1.5
print(k3, k3.dtype)
```
