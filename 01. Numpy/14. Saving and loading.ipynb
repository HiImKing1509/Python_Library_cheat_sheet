# **What is Numpy?**
**NumPy** là thư viện cơ bản cho tính toán khoa học bằng Python. Nó là một thư viện Python cung cấp một đối tượng mảng đa chiều, các đối tượng dẫn xuất khác nhau (chẳng hạn như các mảng và ma trận có mặt nạ) và một loạt các quy trình cho các hoạt động nhanh trên mảng, bao gồm toán học, logic, thao tác hình dạng, sắp xếp, lựa chọn, I / O , các phép biến đổi Fourier rời rạc, đại số tuyến tính cơ bản, các phép toán thống kê cơ bản, mô phỏng ngẫu nhiên và hơn thế nữa.

# **Saving and loading**
> **NumPy** giúp dễ dàng lưu và tải các **ndarrays** ở định dạng nhị phân hoặc văn bản

**Binary $.npy$ format**

```python
arr = np.arange(30).reshape(5, 6)
arr
```

```python
np.save("my_arr", arr)
```

File **$.npy$** được tạo

my_arr.png

Xem qua nội dung file

```python
with open("my_arr.npy", "rb") as f:
    content = f.read()
content
```

Load file vào mảng **numpy**

```python
arr_load = np.load("my_arr.npy", allow_pickle=True)
```

```python
arr_load
```

### **Text format**
> Lưu mảng ở định dạng văn bản

```python
arr
```

```python
np.savetxt("my_arr.csv", arr)
```

Xem qua nội dung file

```python
with open("my_arr.csv", "rt") as f:
    print(f.read())
```

Lưu mảng ở định dạng văn bản format rõ ràng hơn

```python
np.savetxt("my_arr1.csv", arr, delimiter=",")
```

Xem qua nội dung file

```python
with open("my_arr1.csv", "rt") as f:
    print(f.read())
```

Load file vào mảng **numpy**

```python
arr1_load = np.loadtxt("my_arr1.csv", delimiter=",")
```

```python
arr1_load
```

### **Zipped $.npz$ format**
> Có thể lưu mảng trong một tập nén

```python
a = np.arange(24, dtype=np.uint8).reshape(2, 3, 4)
a
```

Lưu file ở định dạng nén

```python
np.savez("my_arr", my_a1=arr, my_a2=a)
```

Load fine nén

```python
my_arr_load = np.load("my_arr.npz")
```

```python
my_arr_load
```

```python
my_arr_load["my_a1"]
```

```python
my_arr_load["my_a2"]
```
